```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# import wanted sdoh 2020 data and columns
data_sdoh_2020 = pd.read_excel("SDOH_2020_COUNTY_cleaned_clinical.xlsx", 
                          sheet_name= 'SDOH_2020',
                          usecols= ['ACS_PCT_HH_NO_INTERNET', 'YEAR', 'REGION', 'COUNTY', 'STATE', 'ACS_PCT_HH_NO_COMP_DEV','ACS_PCT_HH_SMARTPHONE'])
data_sdoh_2020.drop(data_sdoh_2020.index[[0, 1, 2]], inplace=True)
data_sdoh_2020 = data_sdoh_2020[data_sdoh_2020["COUNTY"] != "Valdez-Cordova Census Area"] # dropped because it is missing
data_sdoh_2020["ACS_PCT_HH_NO_COMP_DEV"] = data_sdoh_2020["ACS_PCT_HH_NO_COMP_DEV"].astype(int)
data_sdoh_2020["ACS_PCT_HH_SMARTPHONE"] = data_sdoh_2020["ACS_PCT_HH_SMARTPHONE"].astype(int)
data_sdoh_2020["ACS_PCT_HH_NO_INTERNET"] = data_sdoh_2020["ACS_PCT_HH_NO_INTERNET"].astype(int)



data_sdoh_2020.head()
```

```python
# Count how many missing values exist before filling
missing_count = data_sdoh_2020.isna().sum().sum()

# get location of each missing value
missing_locations = data_sdoh_2020.isna()
rows, cols = np.where(missing_locations)
for r, c in zip(rows, cols):
    print(f"Missing value at row {r}, column '{data_sdoh_2020.columns[c]}'")

# Fill all missing values with "NA"
data1 = data_sdoh_2020.infer_objects(copy=False)

print(f"Filled {missing_count} missing values with 'NA'.")
```

```python
# make tidy for graphing 
df_long_sdoh_2020 = data_sdoh_2020.melt(
    id_vars=["STATE", "COUNTY", "YEAR", "REGION"],
    value_vars=[
        "ACS_PCT_HH_NO_INTERNET",
        "ACS_PCT_HH_SMARTPHONE",
        "ACS_PCT_HH_NO_COMP_DEV"
    ],
    var_name="Outcome",
    value_name="Percent"
)

# Nicer labels
df_long_sdoh_2020["Outcome"] = df_long_sdoh_2020["Outcome"].replace({
    "ACS_PCT_HH_NO_INTERNET": "No Internet Access",
    "ACS_PCT_HH_SMARTPHONE": "Smartphone Access",
    "ACS_PCT_HH_NO_COMP_DEV": "No Computer Device"
})
```

```python
# make tidy for graphing 
df_long_sdoh_2017 = data_sdoh_2017.melt(
    id_vars=["STATE", "COUNTY", "YEAR", "REGION"],
    value_vars=[
        "ACS_PCT_HH_NO_INTERNET",
        "ACS_PCT_HH_SMARTPHONE",
        "ACS_PCT_HH_NO_COMP_DEV"
    ],
    var_name="Outcome",
    value_name="Percent"
)

# Nicer labels
df_long_sdoh_2017["Outcome"] = df_long_sdoh_2017["Outcome"].replace({
    "ACS_PCT_HH_NO_INTERNET": "No Internet Access",
    "ACS_PCT_HH_SMARTPHONE": "Smartphone Access",
    "ACS_PCT_HH_NO_COMP_DEV": "No Computer Device"
})
```

```python
#concact all sdoh for graphing
df_long_sdoh = pd.concat([df_long_sdoh_2020, df_long_sdoh_2019, df_long_sdoh_2018], ignore_index=True)
df_long_sdoh.head()
```

```python
# Filter to internet access outcome and years of interest
df_internet = df_long_sdoh[
    (df_long_sdoh["Outcome"] == "No Internet Access") &
    (df_long_sdoh["YEAR"].isin([2018, 2019, 2020]))
]

# Aggregate counties to state-year means
df_state_year = (
    df_internet
    .groupby(["STATE", "YEAR"], as_index=False)
    .agg(mean_no_internet=("Percent", "mean"))
)
```

```python
fig = px.line(
    df_state_year,
    x="YEAR",
    y="mean_no_internet",
    facet_col="STATE",
    facet_col_wrap=10,   
    markers=True,
    title="Change of Proportion of Popultion with No Internet Access By State (20182020)"
)

fig.update_layout(
    height=1200,      # increase vertical space
    width=2000,       # increase horizontal space
    showlegend=False,
    xaxis_title="Year",
    yaxis_title="Mean % Without Internet Access"
)

fig.update_xaxes(showticklabels=True)
fig.update_yaxes(showticklabels=True)

fig.update_xaxes(
    dtick=1,
    tick0=2018
)

# Make facet titles cleaner
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.for_each_yaxis(lambda yaxis: yaxis.update(title_text=None))


fig.update_yaxes(matches=None)  # allow each state its own y-scale if desired

fig.write_image('state_line_copy.png')
fig.show()
```

```python
fig_sdoh = px.box(
    df_long_sdoh[(df_long_sdoh["Outcome"] == "No Internet Access")],
    x="YEAR",
    y="Percent",
    points="all",     
    hover_data=["STATE", "COUNTY", "YEAR"],
)

fig_sdoh.update_layout(
    title="No Internet Access 2018-2020",
    xaxis_title="Outcome Measure",
    yaxis_title="Percent of Households",
)

fig_sdoh.write_image('box_whisker_internet.png')
fig_sdoh.show()
```
